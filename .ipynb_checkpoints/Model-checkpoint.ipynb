```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
datasetTrain= pd.read_csv('FE_Train.csv')
datasetTest= pd.read_csv('FE_Test.csv')
```

```python
datasetTrain['Survived'].value_counts()
```

```python
datasetTest.head()
```

```python
datasetTrain.corr(method ='pearson')
```

```python
datasetTrain=datasetTrain.drop(['Age', 'SibSp', 'Parch', 'Embarked'], axis=1)
datasetTest=datasetTest.drop(['Age', 'SibSp', 'Parch', 'Embarked'], axis=1)
```

```python
X=datasetTrain.iloc[:, datasetTrain.columns!='Survived']
y=datasetTrain.loc[:, 'Survived']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=2022, shuffle=True)
```

```python
X_train.shape[1]
```

```python
inputs = X_train.shape[1]

clasificador = Sequential()

clasificador.add(Dense(input_dim = inputs, units=15, activation='relu', kernel_initializer='uniform'))

#clasificador.add(Dense(units=12, activation='relu', kernel_initializer='uniform'))

clasificador.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
```

```python
#parametros de optimización
clasificador.compile(optimizer='SGD', loss='binary_crossentropy', metrics=['accuracy'])

#parametros de entrenamiento.
clasificador.fit(X_train, y_train, batch_size=25, epochs=150)
```

```python
clasificador.save('12_sig_out.h5')
```

```python
new_model = keras.models.load_model('./Configuraciones/12_sig_out.h5')
```

```python
y_preds = new_model.predict(datasetTest)
#y_preds
```

```python
y_preds = (y_preds >= 0.5)
```

```python
dfUpload=pd.read_csv('test.csv')
dfUpload['Survived']=y_preds
```

```python
dfUpload=dfUpload.drop(['Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1)
dfUpload["Survived"]=dfUpload["Survived"].replace(False, 0, regex=True)
dfUpload["Survived"]=dfUpload["Survived"].replace(True, 1, regex=True)
dfUpload.head()
```

```python
dfUpload.to_csv(r'./Resultados de los modelos/ResultFinal.csv', index=False)
```
